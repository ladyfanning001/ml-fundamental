```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test_scores.csv')
df
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    rate = 0.0002  # lebih kecil biar stabil
    n = len(x)
    plt.scatter(x, y, color='red', marker='+', linewidth=5)

    for i in range(70000):  # iterasi aku kecilin juga
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * bd
        plt.plot(x, y_predicted, color='green')
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))
        if np.isnan(m_curr) or np.isnan(b_curr) or np.isnan(cost) or np.isnan(cost) or np.isinf(cost):
            print(f"Stop di iterasi {i} karena muncul NaN")
            break

        plt.plot(x, y_predicted, color='green')
        print("m {}, b {}, cost {} iteration {}".format(m_curr, b_curr, cost, i))

    plt.show()
```

```python
x = df['math'].values
y = df['cs'].values
gradient_descent(x, y)
```
