```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
#using pandas to create dummy variables
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
#merged df
final = pd.concat([df,dummies], axis = 'columns')
final
```

```python
#drop one of dummies column
final = final.drop(['Car Model'], axis = 'columns')
final
```

```python
#dummy variable trap, delete one column (co-linear)
final = final.drop('Audi A5', axis = 'columns')
final
```

```python
x = final.drop("Sell Price($)", axis = 'columns')
x
```

```python
y = final.loc[:,"Sell Price($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[45000, 4, 1, 0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x, y)
```
